```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(11) # set seed to roll no. 23091011
C = np.random.randint(-3, 4,6)
C # coefficients
```

```python
# so the polynomial is -2 - 3x - 2x^3 + 2x^4 + x^5
X = np.arange(-5, 5+0.05, 0.05) # set the range
plt.plot(X, (C[0] + C[1]*X + C[2]*X**2 + C[3]*X**3 + C[4]*X**4 + C[5]*X**5), color="blue")
plt.annotate("local minima", xy = (-4.7, -1500), xytext = (-4, -1000), arrowprops = dict(facecolor = "black", shrink = 0))
plt.annotate("local maxima", xy = (4.8, 4100), xytext = (1.8, 3800), arrowprops = dict(facecolor = "black", shrink = 0))
```
